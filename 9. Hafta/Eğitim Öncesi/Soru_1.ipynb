YetGen için bir kahve otomatı yapmak istiyoruz. Bu kahve otomatında aşağıdaki adımları gerçekleştirmesini bekliyoruz. Bu işlemler:

Kahve Ekle
Fiyat Güncelleme
Miktarı Değiştirme (ekle/kaldır)
Otomatı Açma
Otomatı Kapatma

Sorumuzun bu bölümünde sizlerden kahve otomatını oluşturmanızı ve otomat açma, kapama işlemlerini gerçekleştirmenizi bekliyoruz.

```python
fiyat = {
    "espresso": [20,25,27],
    "americano": [15,20,35],
    "latte": [40,45,50]
}

class otomat():
    
    
    def __init__(self, otomat_durum="kapalı", kahve_listesi=["espresso", "americano", "latte"], miktar="küçük",
                 seker=0):
        self.otomat_durum = otomat_durum
        self.kahve_listesi = kahve_listesi
        self.miktar = miktar
        self.fiyat = fiyat
        
# YetNot: Bu bölümde otomat_ac isimli bir fonksiyon oluşturup, ekrana "Otomat Açılıyor" yazdırıp, otomatının durumunu açık hale getirmelisin.
    def otomat_ac(self):
        if self.otomat_durum == "kapalı":
            print("Otomat Açılıyor.")
            self.otomat_durum = "açık"
        else:
            print("Otomat halihazırda açık.")

    def otomat_kapat(self):
        if self.otomat_durum == "kapalı":
            print("Önce otomatı açınız.")
        else:
            if (self.otomat_durum == "kapalı"):
                print("Otomat zaten kapalı....")
            else:
                print("Otomat kapanıyor...")
                self.otomat_durum = "kapalı"


otomat = otomat()

print("""
Kahve Otomatı
1. Otomatı aç
2. Otomatı kapat
3. Menüyü yazdır
4. Kahve ekle
5. Kahvenizi seçin
6. Fiyat güncelle
Çıkmak için 'q' ya basın.""")



while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("Afiyet olsun!!!")
        break
    elif (islem == "1"):
        otomat.otomat_ac()
    elif (islem == "2"):
        otomat.otomat_kapat()
    elif islem == "3":
        print("Mevcut menü: ", otomat.kahve_listesi)
    elif islem == "4":
        yeni_kahve = input("Eklemek istediğiniz kahvenin adını girin: ")
        if yeni_kahve not in otomat.kahve_listesi:
            otomat.kahve_listesi.append(yeni_kahve)
            yeni_fiyat = float(input(f"{yeni_kahve} için fiyat girin: "))
            otomat.fiyat[yeni_kahve] = yeni_fiyat
            print(f"{yeni_kahve} menüye eklendi.")
        else:
            print(f"{yeni_kahve} zaten menüde mevcut.")
    elif islem == "5":
        secilen_kahve = input("Seçtiğiniz kahveyi girin: ")
        if secilen_kahve in otomat.kahve_listesi:
            print(f"Seçtiğiniz kahve: {secilen_kahve}, Fiyatı: {otomat.fiyat[secilen_kahve]} TL")
        else:
            print("Seçtiğiniz kahve menüde yok.")
    elif islem == "6":
        kahve = input("Fiyatını güncellemek istediğiniz kahveyi girin: ")
        if kahve in otomat.kahve_listesi:
            yeni_fiyat = float(input(f"Yeni fiyatı girin ({kahve}): "))
            otomat.fiyat[kahve] = yeni_fiyat
            print(f"{kahve} için yeni fiyat: {yeni_fiyat} TL")
        else:
            print("Güncellemek istediğiniz kahve menüde yok.")
    else:
        print("Geçersiz işlem, tekrar deneyin.")
```
